```python
# !pip install pandas
# !pip install numpy
import pandas as pd
import numpy as np

df=pd.read_csv('automobile.csv')
df.head()
```

```python
# preprocessing

print(df.isnull().sum())
```

```python
# handling null values
# print(df['normalized-losses'].mean())
df['normalized-losses'].fillna(df['normalized-losses'].mean(),inplace=True)
df['num-of-doors'].fillna(df['num-of-doors'].mode(),inplace=True)
```

```python
df
```

```python
# plot
# !pip install matplotlib
import matplotlib.pyplot as plt
x=np.array([i for i in range(202)])
y=np.array(df['price'])

plt.plot(x,y,'o')
plt.xlabel('Number')
plt.ylabel('Price')
plt.title('Prices of cars')
plt.show()
ans=[]
for i in df['price']:
    print(i)
    if i>=10000:
        ans.append(1)
    else:
        ans.append(0)
actaul=pd.DataFrame(ans)
df.drop(['price'],axis=1,inplace=True)
print(actaul)

```

```python
df
actaul.head(50)
```

```python
# !pip install -U scikit-learn

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split 
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df,actaul,test_size=0.2,random_state=42)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
from sklearn.preprocessing import LabelEncoder
#encode the categorical data
le=LabelEncoder()

for i in df:
    df[i]=le.fit_transform(df[i])



model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)

model2=RandomForestClassifier()

model2.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import classification_report
dtree=(classification_report(ytest,model.predict(xtest)))
rforest=(classification_report(ytest,model2.predict(xtest)))

print(dtree)
print(rforest)


```

```python

import matplotlib.pyplot as plt

dtree=dtree.split()
# print(dtree)
xaxis=[
    "",
    dtree[14],
    dtree[0],
    dtree[1],
    dtree[2]
]
yaxis=[
    '0',
    dtree[15],
    dtree[5],
    dtree[6],
    dtree[7], 
]
plt.bar(xaxis,yaxis)
plt.title('DecisionTreeClassifier')
plt.show()


rforest=rforest.split()
# print(rforest)
xaxis=[
    "",
    rforest[14],
    rforest[0],
    rforest[1],
    rforest[2]
]
print(xaxis)
yaxis=[
    '0',
    rforest[15],
    rforest[5],
    rforest[6],
    rforest[7]
]
print(yaxis)
plt.bar(xaxis,yaxis,color='red')
plt.title('RandomForestClassFier')

plt.show()
```

```python
from sklearn.metrics import classification_report
dtree=(classification_report(ytest,model.predict(xtest)))
rforest=(classification_report(ytest,model2.predict(xtest)))
import matplotlib.pyplot as plt

dtree=dtree.split()
rforest=rforest.split()
X = [
    dtree[14],
    dtree[0],
    dtree[1],
    dtree[2],
]
dtreeee = [
    dtree[15],
    dtree[5],
    dtree[6],
    dtree[7], 
]
rffff = [
    rforest[15],
    rforest[5],
    rforest[6],
    rforest[7]
]
  
X_axis = np.arange(len(X)) 
  
plt.bar(X_axis+0.2 , dtreeee, 0.4, label = 'DescisonTree') 
plt.bar(X_axis-0.2, rffff, 0.4, label = 'RadomForest ') 
  
plt.xticks(X_axis, X) 
plt.xlabel("Metrics") 
plt.ylabel("score") 
plt.title("DescisonTree vs RadomForest") 
plt.legend() 
plt.show() 

```
