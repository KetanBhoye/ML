# Dataset Loading 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('automobile.csv')
```

```python
df = df.dropna()
```

# EDA and Preprocessing

```python
df.describe().T
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
category = ['make', 'fuel-type', 'aspiration', 'engine-location' ,'fuel-system', 'body-style','num-of-doors','drive-wheels','engine-type','horsepower','num-of-cylinders','bore','stroke']
label_encoders = {}
for feature in category:
    label_encoders[feature] = LabelEncoder()
    df.loc[:,feature] = label_encoders[feature].fit_transform(df[feature])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df)
```

```python
scaled_data
```

```python
df_scaled = pd.DataFrame(data=scaled_data, columns=df.columns)
```

```python
x = df_scaled.drop(['price'], axis=1)
```

```python
y = df_scaled['price']
```

# Apply ML Model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, y_train, x_test, y_test  = train_test_split(x,y, test_size=0.5,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train)
```

# Evaluation and Accuracy

```python
y_pred = regressor.predict(x_train)
```

```python
y_pred
```

```python
regressor.score(y_pred,y_test)
```
